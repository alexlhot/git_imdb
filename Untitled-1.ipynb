```python
import json
import pathlib

dico = json.loads(pathlib.Path("pokemon.json").read_text())
```

```python
class Clumper:
    def __init__(self, blob):
        self.blob = blob

    def keep(self, func):
        return [d for d in self.blob if func(d)]
```

```python
Clumper(dico).keep(lambda d: 'Grass' in d['type'])
```

```python
[print(d) for d in dico if 'Grass' in d['type']]
```
